## Objective

Learn how to write effective prompts for Large Language Models.

## What is Prompt Engineering?

Prompt Engineering is the process of designing prompts to get better AI responses.

## Zero-Shot Prompting

The model is given only a task.

Example:

Translate English to Tamil.

```python
prompt = "Translate Hello into Tamil"

print(prompt)
```

## One-Shot Prompting

One example is provided before asking the task.

## One-Shot Prompting

One example is provided before asking the task.

## Few-Shot Prompting

Multiple examples are provided before the actual task.

## Role Prompting

Assigning a role to the model.

Example:

You are a Python Expert.

```python
role = "You are a Python Expert"

print(role)
```

## Chain of Thought Prompting

The model is encouraged to think step by step.

## Structured Prompting

Requests output in a specific format.

Example:

Return output as JSON.

```python
prompt = """
Return student information in JSON format
"""

print(prompt)
```

## Best Practices

- Be clear
- Be specific
- Provide examples
- Define output format
