# Day 4: Transformers

## Objective
To understand Transformer architecture and its role in modern Large Language Models.

## Introduction

Transformers are deep learning models introduced in the paper "Attention Is All You Need" (2017).

They are widely used in:
- GPT
- BERT
- Gemini
- Claude

## Why Transformers?

Limitations of RNNs and LSTMs:

- Sequential Processing
- Slow Training
- Difficulty handling long-term dependencies

Transformers solve these issues using Attention Mechanisms.
