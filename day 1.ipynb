```python
fruits = ["Apple", "Banana", "Orange"]

print(fruits)
print(fruits[0])
```

```python
numbers = {1, 2, 3, 3, 4, 5}

print(numbers)
```

## Functions

## Object-Oriented Programming (OOP)

OOP helps organize code using classes and objects.

### Class and Object

```python
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def display(self):
        print("Name:", self.name)
        print("Marks:", self.marks)

s1 = Student("Rithish", 90)
s1.display()
```

### Inheritance
Inheritance allows a class to acquire properties and methods from another class.

```python
class Person:
    def speak(self):
        print("I can speak")

class Student(Person):
    pass

obj = Student()
obj.speak()
```

### Polymorphism
Polymorphism allows different objects to use the same method name.

```python
class Dog:
    def sound(self):
        print("Bark")

class Cat:
    def sound(self):
        print("Meow")

for animal in [Dog(), Cat()]:
    animal.sound()
```

### Encapsulation
Encapsulation hides internal data and provides controlled access.

```python
class BankAccount:
    def __init__(self):
        self.__balance = 1000

    def show_balance(self):
        print(self.__balance)

account = BankAccount()
account.show_balance()
```

## Exception Handling
Exception handling is used to manage runtime errors.

```python
try:
    result = 10 / 0
except ZeroDivisionError:
    print("Cannot divide by zero")
```

## File Handling
File handling is used to read and write files in Python.
